```python
nba = pd.read_csv('https://dadosdatascience.netlify.com/nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')
nba
```

```python
nba.sort_values(by = 'Salary', ascending= False)
```

```python
nba['anual_salary'] = nba.Salary * 12 # adicionando uma nova coluna
nba['anual_salary'] = nba['anual_salary'].dropna().map('${:,.2f}'.format) # formatando o valor final formato moeda($)
nba[['Name', 'Salary','anual_salary']]

```

```python
# Adicionando uma nova coluna ID
nba.insert(0, column='ID', value = range(1, 1+len(nba)))
nba.head()
```

```python
# Gráfico com os 30 maiores salários da nba
from matplotlib import pyplot as plt

salary = nba.sort_values(by = 'Salary', ascending= False).head(20)
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(salary.Name, round(salary.Salary/1000000, 1), color = 'Aqua',edgecolor = 'DarkBlue', label='$M')
# Remove axes splines

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(visible = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('20 Biggest NBA Salaries',
             loc ='center', color='DarkRed' )
 
# Add Text watermark
fig.text(0.9, 0.15, 'NBA SPORTS', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
ax.legend(fontsize=15,loc = 'center right')
plt.show()
```

```python
nba.groupby('Team').size() # numero de atletas por time
#team = nba.groupby('Team').sum('Salary').sort_values(by='Salary', ascending = False)
#team
team = nba.groupby("Team").agg({'Salary':'sum'}).sort_values(by='Salary', ascending = False) # salário dos jogadores somados por time
#team['Salary']
team
```

```python
fig, ax = plt.subplots(figsize =(20, 12))

# Horizontal Bar Plot
ax.barh(team.index, round(team.Salary/1000000, 1), color = 'Aqua',edgecolor = 'DarkBlue', label='$M')
# Remove axes splines

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(visible = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('NBA Total Salaries by Teams ',
             loc ='center', color='DarkRed' )
 
# Add Text watermark
fig.text(0.9, 0.15, 'NBA SPORTS', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
ax.legend(fontsize=15,loc = 'center right')
plt.show()
```
