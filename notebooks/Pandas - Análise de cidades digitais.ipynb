Dados obtidos do Portal Brasileiro de Dados Abertos: (http://dados.gov.br/dataset/cidades-digitais)

```python
import numpy as np
import pandas as pd
```

```python
cidadesDigitais = pd.read_csv('dados/cidades_digitais.csv', sep = ';', encoding = 'latin-1') # Importando arquivo com dados
cidadesDigitais.head()
```

```python
cidadesDigitais.count() # Quantificando dados tabela
```

```python
del cidadesDigitais['IBGE'] # Deletando coluna IBGE
cidadesDigitais.head()
```

```python
cidadesUF = cidadesDigitais.groupby('UF') # Agrupa os dados pela coluna UF
cidadesUF.groups # verifica os grupos formados
len(cidadesUF.groups)
ordenaUF = cidadesUF.size() # armazena os dados agrupados por UF na variavel ordenaUF
```

```python
ordenaUF.sort_values(ascending = False) # Ordena os grupos por ordem decrescente
```

```python
ordenaUF # estrutura não é modificada a não ser use flag encoder
ordenaUF.sort_values(ascending = False, inplace = True)  # com encoder True se modifica o dataframe original
ordenaUF
```

```python
cidadesUF2 = cidadesDigitais.groupby(['UF', 'STATUS'])
ordenaDados = cidadesUF2.size()
ordenaDados.sort_values(ascending = False, inplace = True)
ordenaDados
```

```python
cidadesUF3 = cidadesDigitais.set_index('STATUS') # Transforma a coluna STATUS em indice num novo dataframe
cidadesUF3 = cidadesUF3.groupby(level = 'STATUS') # Agrupa os dados pelo indice STATUS
cidadesUF3.size() 
```

```python
cidadesUF3.head()
```

```python
cidadesUF4 = cidadesDigitais.set_index(['UF', 'STATUS']) # Transforma duas colunas em indices
cidadesUF4 = cidadesUF4.groupby(level = ['UF', 'STATUS']) 
cidadesUF4.sum()
```

```python
cidadesUF4.aggregate(np.sum) # outra forma de agregar pela soma dos dados agrupados
cidadesUF4.aggregate([np.sum, np.mean, np.size]) # dados agregados em media, soma, tamanho
```

```python
cidadeUF4 = cidadesUF4.aggregate([np.sum, np.mean, np.size]).apply(lambda x: round(x)) # Dados agrupados como anteriormente mas com a função round() aplicada sobre os valores
cidadeUF4
#cidadeUF4.POPULAÇÃO.sort_values('size', ascending = False)
#cidadeUF4.PONTOS_ATENDIDOS.sort_values('size', ascending = False)
#cidadeUF4['POPULAÇÃO'][['sum','mean']]
#cidadeUF4.POPULAÇÃO.mean
```
