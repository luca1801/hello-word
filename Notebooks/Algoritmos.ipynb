# **Searching(Algoritmos de Busca)**

Searching are algoritms that you can use to find a particular piece of data in a sorted array.

**1. Linear Search:** simply loops through the items in the array, looking for the target item. Unlike binary search and interpolation search, linear search works on linked lists, where you cannot easily jump from one part of the list to another, as you
can in an array.
